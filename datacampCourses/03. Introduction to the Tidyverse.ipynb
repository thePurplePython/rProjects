## _1. Data Wrangling_:
-  Tidyverse:
    -  collection of R tools for transforming and visualizing data
-  dplyr pipes:
    -  %>% (**used when chaining functions together**)
-  dplyr verbs:
    -  filter
    -  head
    -  arrange (**sorts data on col; default ascending order**)
    -  arrange(desc(...)) (**sorts data on col in descending order**)
    -  mutate (**change or add variable [col]**)

#### install / import / view gapminder dataset

```r
#install.packages("gapminder")
library(gapminder)
head(gapminder)
```

#### install / import dplyr package

```r
#install.packages("dplyr")
library(dplyr)
```

#### filter verb

```r
gapminder %>%
  filter(year == 1957) %>%
  head(10)
```

```r
gapminder %>%
  filter(country == "China", year == 2002)
```

#### arrange verb [ascending order]

```r
gapminder %>%
  arrange(lifeExp) %>%
  head(10)
```

#### arrange verb [descending order]

```r
gapminder %>%
  arrange(desc(lifeExp)) %>%
  head(10)
```

#### chain multiple verbs

```r
gapminder %>%
  filter(year == 1957) %>%
  arrange(desc(pop)) %>%
  head(10)
```

```r
gapminder %>%
  filter(year == 2007) %>%
  mutate(lifeExpMonths = 12 * lifeExp) %>%
  arrange(desc(lifeExpMonths)) %>%
  head(10)
```

#### mutate verb

```r
# update existing col
gapminder %>%
  mutate(lifeExp = 12 * lifeExp) %>%
  head(10)

# add new calculated col
gapminder %>%
  mutate(lifeExpMonths = 12 * lifeExp) %>%
  head(10)
```

## _2. Data Visualization_:
-  ggplot2 visual aesthetics:
    -  scatter plots => **geom_point()**:
        -  log scale:
            -  scale where each fixed distance represents a multiplication of the value
            -  improves linear and plotting representation
        -  categorical color labels
        -  size
        -  faceting:
            -  sub plots by variable

#### install / import ggplot2

```r
#install.packages("ggplot2")
library(ggplot2)
```

#### filter + visual

```r
gapminder_1952 <- gapminder %>%
  filter(year == 1952)

ggplot(gapminder_1952, aes(x = pop, y = gdpPercap)) +
  geom_point()
ggplot(gapminder_1952, aes(x = pop, y = lifeExp)) +
  geom_point()
```

#### x axis log scale

```r
ggplot(gapminder_1952, aes(x = pop, y = lifeExp)) +
  geom_point() +
  scale_x_log10()
```

#### x & y axis log scale

```r
ggplot(gapminder_1952, aes(x = pop, y = gdpPercap)) +
  geom_point() +
  scale_x_log10() +
  scale_y_log10()
```

#### color labels

```r
ggplot(gapminder_1952, aes(x = pop, y = lifeExp, color = continent)) +
  geom_point() +
  scale_x_log10()
```

#### size

```r
ggplot(gapminder_1952, aes(x = pop, y = lifeExp, color = continent, size = gdpPercap)) +
  geom_point() +
  scale_x_log10()
```

#### faceting

```r
ggplot(gapminder_1952, aes(x = pop, y = lifeExp)) +
  geom_point() +
  scale_x_log10() +
  facet_wrap(~ continent) # plotting by continent values
```

```r
ggplot(gapminder, aes(x = gdpPercap, y = lifeExp, color = continent, size = pop)) +
  geom_point() +
  scale_x_log10() +
  facet_wrap(~ year) # plotting by continent years
```

## _3. Grouping and Summarizing_:
-  verbs:
    -  summarize (**entire dataset aggregation; returns 1 row**)
    -  group_by

#### summarize

```r
gapminder %>%
  summarize(medianLifeExp = median(lifeExp))

gapminder %>%
  filter(year == 1957) %>%
  summarize(medianLifeExp = median(lifeExp))
```

#### multiple summarize

```r
gapminder %>%
  filter(year == 1957) %>%
  summarize(medianLifeExp = median(lifeExp),
            maxGdpPercap = max(gdpPercap))
```

#### group by

```r
gapminder %>%
  group_by(year) %>%
  summarize(medianLifeExp = median(lifeExp),
            maxGdpPercap = max(gdpPercap))

gapminder %>%
  filter(year == 1957) %>%
  group_by(continent) %>%
  summarize(medianLifeExp = median(lifeExp),
            maxGdpPercap = max(gdpPercap))
```

#### multiple group by

```r
gapminder %>%
  group_by(continent, year) %>%
  summarize(medianLifeExp = median(lifeExp),
            maxGdpPercap = max(gdpPercap))
```

#### visualize group by

```r
by_year <- gapminder %>%
  group_by(year) %>%
  summarize(medianLifeExp = median(lifeExp),
            maxGdpPercap = max(gdpPercap))

ggplot(by_year, aes(x = year, y = medianLifeExp)) +
  geom_point() +
  expand_limits(y = 0) # scales y axis to include 0
```

#### visualize multiple group by

```r
by_year_continent <- gapminder %>%
  group_by(continent, year) %>%
  summarize(medianGdpPercap = median(gdpPercap))

ggplot(by_year_continent, aes(x = year, y = medianGdpPercap, color = continent)) +
  geom_point() +
  expand_limits(y = 0)
```

#### visualize summary aggregation

```r
by_continent_2007 <- gapminder %>%
  filter(year == 2007) %>%
  group_by(continent) %>%
  summarize(medianGdpPercap = median(gdpPercap),
            medianLifeExp = median(lifeExp))

ggplot(by_continent_2007, aes(x = medianGdpPercap, y = medianLifeExp, color = continent)) +
  geom_point()
```

## _4. Types of Visualizations_:
-  line plots => show change over time (**geom_line()**)
-  bar plots => compare stats on several discrete categories (**geom_col()**)
-  histograms => describe binned distribution of 1D numeric variable (**geom_histogram()**)
-  boxplots => compare distribution of numeric variable among several categories (**geom_boxplot()**)

#### single line plot

```r
by_year <- gapminder %>%
  group_by(year) %>%
  summarize(medianGdpPercap = median(gdpPercap))

ggplot(by_year, aes(x = year, y = medianGdpPercap)) +
  geom_line() +
  expand_limits(y = 0)
```

#### multiple line plot

```r
by_year_continent <- gapminder %>%
  group_by(year, continent) %>%
  summarize(medianGdpPercap = median(gdpPercap))

ggplot(by_year_continent, aes(x = year, y = medianGdpPercap, color = continent)) +
  geom_line() +
  expand_limits(y = 0)
```

#### bar plot

```r
by_continent <- gapminder %>%
  filter(year == 1952) %>%
  group_by(continent) %>%
  summarize(medianGdpPercap = median(gdpPercap))

ggplot(by_continent, aes(x = continent, y = medianGdpPercap)) +
  geom_col()

oceania_1952 <- gapminder %>%
  filter(continent == "Oceania", year == 1952)

ggplot(oceania_1952, aes(x = country, y = gdpPercap)) +
  geom_col()
```

#### histogram plot

```r
gapminder_1952 <- gapminder %>%
  filter(year == 1952)

ggplot(gapminder_1952, aes(x = pop)) +
  geom_histogram()
```

#### log scale histogram plot

```r
gapminder_1952 <- gapminder %>%
  filter(year == 1952)

ggplot(gapminder_1952, aes(x = pop)) +
  geom_histogram() +
  scale_x_log10()
```

#### boxplot

```r
gapminder_1952 <- gapminder %>%
  filter(year == 1952)

ggplot(gapminder_1952, aes(x = continent, y = gdpPercap)) +
  geom_boxplot() +
  scale_y_log10()
```

#### add title

```r
gapminder_1952 <- gapminder %>%
  filter(year == 1952)

ggplot(gapminder_1952, aes(x = continent, y = gdpPercap)) +
  geom_boxplot() +
  scale_y_log10() +
  ggtitle("Comparing GDP per capita across continents")
```
